```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))

def relu(x):
    return max(0, x)

def leaky_relu(x):
    return max(0.1 * x, x)
```

```python
fig = plt.figure(figsize=(20,10))
fig.tight_layout()
gs = fig.add_gridspec(2, 2)  

lower, upper = (-20, 20)
x = random.uniform(lower, upper)
xLine = np.linspace(lower, upper, 100000)
y1Line = [sigmoid(x) for x in xLine]
y2Line = [tanh(x) for x in xLine]
y3Line = [relu(x) for x in xLine]
y4Line = [leaky_relu(x) for x in xLine]

a1 = fig.add_subplot(gs[0, 0])
a1.plot(xLine, y1Line)
a1.set_title(f'sigmoid({np.round(x, decimals=4)}) = {sigmoid(x)}')
a1.plot(x, sigmoid(x), marker='o', color='red')
a1.set_ylabel('sigmoid(x)')
a1.set_xlabel('x')

a2 = fig.add_subplot(gs[0, 1])
a2.plot(xLine, y2Line)
a2.set_title(f'tanh({np.round(x, decimals=4)}) = {tanh(x)}')
a2.plot(x, tanh(x), marker='o', color='red')
a2.set_ylabel('tanh(x)')
a2.set_xlabel('x')

a3 = fig.add_subplot(gs[1, 0])
a3.plot(xLine, y3Line)
a3.set_title(f'relu({np.round(x, decimals=4)}) = {relu(x)}')
a3.plot(x, relu(x), marker='o', color='red')
a3.set_ylabel('relu(x)')
a3.set_xlabel('x')

a4 = fig.add_subplot(gs[1, 1])
a4.plot(xLine, y4Line)
a4.set_title(f'leaky_relu({np.round(x, decimals=4)}) = {leaky_relu(x)}')
a4.plot(x, leaky_relu(x), marker='o', color='red')
a4.set_ylabel('leaky_relu(x)')
a4.set_xlabel('x')

plt.show()
```
